# Teste t, Cohens, Magnitude

```python
import numpy as np
from scipy import stats

# -------------------------------
# Função para Cohen's d
# -------------------------------

def cohens_d(x, y):
    nx = len(x)
    ny = len(y)

    pooled_sd = np.sqrt(
        ((nx-1)*np.var(x, ddof=1) + (ny-1)*np.var(y, ddof=1))
        /(nx+ny-2)
    )

    d = (np.mean(x)-np.mean(y))/pooled_sd

    return d


# -------------------------------
# Magnitude
# -------------------------------

def magnitude(d):
    d = abs(d)

    if d < 0.20:
        return "Insignificante"
    elif d < 0.50:
        return "Pequeno"
    elif d < 0.80:
        return "Médio"
    else:
        return "Grande"


# -------------------------------
# Comparação completa
# -------------------------------

def comparar(nome1, grupo1, nome2, grupo2):

    grupo1 = np.array(grupo1)
    grupo2 = np.array(grupo2)

    print("="*70)
    print(f"{nome1} VS {nome2}")
    print("="*70)

    print(f"N1 = {len(grupo1)}")
    print(f"N2 = {len(grupo2)}")

    print(f"Média 1 = {grupo1.mean():.2f}")
    print(f"Média 2 = {grupo2.mean():.2f}")

    # --------------------
    # Levene
    # --------------------

    stat_lev, p_lev = stats.levene(grupo1, grupo2)

    igual = p_lev > 0.05

    # --------------------
    # t Student
    # --------------------

    t, p = stats.ttest_ind(
        grupo1,
        grupo2,
        equal_var=igual
    )

    # --------------------
    # Graus de liberdade
    # --------------------

    if igual:
        gl = len(grupo1)+len(grupo2)-2
    else:
        gl = np.nan

    # --------------------
    # IC95% diferença
    # --------------------

    media_diff = grupo1.mean()-grupo2.mean()

    se = np.sqrt(
        np.var(grupo1, ddof=1)/len(grupo1)
        +
        np.var(grupo2, ddof=1)/len(grupo2)
    )

    ic = stats.t.interval(
        0.95,
        df=gl,
        loc=media_diff,
        scale=se
    )

    # --------------------
    # Cohen d
    # --------------------

    d = cohens_d(grupo1, grupo2)

    print()

    print(f"Levene p = {p_lev:.2f}")

    print(f"t = {t:.2f}")

    print(f"gl = {gl}")

    print(f"p = {p:.2f}")

    print(f"Diferença entre médias = {media_diff:.2f}")

    print(f"IC95% = ({ic[0]:.2f}; {ic[1]:.2f})")

    print(f"Cohen d = {abs(d):.2f}")

    print(f"Magnitude = {magnitude(d)}")

    if p < 0.05:
        print("Diferença significativa")
    else:
        print("Não significativa")

    print()


# ------------------------------------------------
# Comparações do artigo
# ------------------------------------------------

comparar(
    "Impresso Controle",
    impresso_controle,
    "Impresso Experimental",
    impresso_experimental
)

comparar(
    "Digital Controle",
    digital_controle,
    "Digital Experimental",
    digital_experimental
)

comparar(
    "Impresso Controle",
    impresso_controle,
    "Digital Controle",
    digital_controle
)

comparar(
    "Impresso Experimental",
    impresso_experimental,
    "Digital Experimental",
    digital_experimental
)
```
